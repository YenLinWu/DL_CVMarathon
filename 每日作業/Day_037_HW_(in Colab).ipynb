# Step 4  

輸出層解碼 : 

```python
#輸出層解碼
def process_predicts(predicts):
    p_classes = predicts[0, :, :, 0:20]
    C = predicts[0, :, :, 20:22]
    coordinate = predicts[0, :, :, 22:]

    p_classes = np.reshape(p_classes, (7, 7, 1, 20))
    C = np.reshape(C, (7, 7, 2, 1))

    # 對應相乘，產生𝑝(〖𝐶𝑙𝑎𝑠𝑠〗_𝑗 |𝑜𝑏𝑗𝑒𝑐𝑡)*𝑃(𝑜𝑏𝑗𝑒𝑐𝑡)
    P = C * p_classes
    index = np.argmax(P)

    index = np.unravel_index(index, P.shape)

    class_num = index[3]

    coordinate = np.reshape(coordinate, (7, 7, 2, 4))

    max_coordinate = coordinate[index[0], index[1], index[2], :]

    xcenter = max_coordinate[0]
    ycenter = max_coordinate[1]
    w = max_coordinate[2]
    h = max_coordinate[3]

    xcenter = (index[1] + xcenter) * (448/7.0)
    ycenter = (index[0] + ycenter) * (448/7.0)

    w = w * 448
    h = h * 448

    xmin = xcenter - w/2.0
    ymin = ycenter - h/2.0

    xmax = xmin + w
    ymax = ymin + h

    return xmin, ymin, xmax, ymax, class_num
```

# Step 5   

設定 YOLO 的超參數 : 

```python
common_params = {'image_size':448, 
                 'num_classes':20, 
                 'batch_size':1}

net_params = {'cell_size':7, 'boxes_per_cell':2, 'weight_decay':0.0005, 
              'class_scale':2.0, 'object_scale':1.0, 'noobject_scale':0.5, 'coord_scale':5.0}


net = YoloNet(common_params, net_params, test=False)

# Tensorflow 如果想從外部傳入資料,需要使用 tf.placeholder()，然後以這種形式傳輸數據， sess.run(***,feed_dict={input: **})
# 傳值的工作交給 sess.run()，需要傳入的值放在 feed_dict={}，一一對應每一個 input, placeholder和 feed_dict={}是綁定在一出現的。 
image = tf.placeholder( tf.float32, (1,448,448,3) ) 
predicts = net.inference(image)
```

通過這些參數，可以看到模型各層的組成，數據經過每一層後輸出的數據維度，及參數數量是怎麼算出來的。

```python
import tensorflow.contrib.slim as slim
def model_summary():
    model_vars = tf.trainable_variables()
    slim.model_analyzer.analyze_vars(model_vars, print_info=True)

model_summary()
```
